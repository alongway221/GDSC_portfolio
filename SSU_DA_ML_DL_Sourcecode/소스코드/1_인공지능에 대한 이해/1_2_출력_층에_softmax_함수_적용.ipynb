## 1. 시그모이드(Sigmoid) 함수

```python
from math import exp, tanh, log

i1, i2 = .05, .10
t1, t2 =   0,   1 #목표 값을 0과 1로 변경

w1, w3 = .15, .25
w2, w4 = .20, .30
b1, b2 = .35, .35

w5, w7 = .40, .50
w6, w8 = .45, .55
b3, b4 = .60, .60

for epoch in range(2000000):

	#print('epoch = %d' %epoch)

	h1 = i1*w1 + i2*w2 + 1*b1
	h2 = i1*w3 + i2*w4 + 1*b2
	h1 = 1/(1+exp(-h1)) #h1노드에 순전파 시그모이드 활성화 함수 적용
	h2 = 1/(1+exp(-h2)) #h2노드에 순전파 시그모이드 활성화 함수 적용

	o1 = h1*w5 + h2*w6 + 1*b3
	o2 = h1*w7 + h2*w8 + 1*b4
	o1m = o1 - max(o1, o2) #출력 층의 활성화 함수를 소프트맥스로 변경, o1과 o2 중 max값을 출력
	o2m = o2 - max(o1, o2) #출력 층의 활성화 함수를 소프트맥스로 변경, o1과 o2 중 max값을 출력
	o1 = exp(o1m)/(exp(o1m)+exp(o2m)) #출력 층의 활성화 함수를 소프트맥스로 변경
	o2 = exp(o2m)/(exp(o1m)+exp(o2m)) #출력 층의 활성화 함수를 소프트맥스로 변경

	#print(' o1,  o2  = %6.3f, %6.3f' %(o1, o2))

	E = -t1*log(o1) + -t2*log(o2) #오차 계산을 크로스 엔트로피 오차 형태의 수식으로 변경. 소프트맥스 활성화 함수는 크로스 엔트로피 오차와 같이 사용
	if E < 0.0001:                #for문을 빠져나가는 오차 값을 0.0001로 변경. 여기서 사용하는 값의 크기에 따라 학습의 정확도와 학습 시간이 결정됨
		break

	o1b, o2b = o1 - t1, o2 - t2 #소프트맥스 함수의 역전파 오차 계산. 크로스 엔트로피 함수와 같이 사용될 때 역전파 시 소프트맥스 함수를 역으로 거쳐 전파되는 오차가 예측 값과 목표 값의 차가 됨
	# nothing for softmax + cross entropy error

	h1b, h2b = o1b*w5+o2b*w7, o1b*w6+o2b*w8
	h1b, h2b = h1b*h1*(1-h1), h2b*h2*(1-h2) #h1b, h2b노드에 역전파 시그모이드 활성화 함수 적용

	w1b, w3b = i1*h1b, i1*h2b
	w2b, w4b = i2*h1b, i2*h2b
	b1b, b2b = 1*h1b, 1*h2b
	w5b, w7b = h1*o1b, h1*o2b
	w6b, w8b = h2*o1b, h2*o2b
	b3b, b4b = 1*o1b, 1*o2b

	lr = 0.01
	w1, w3 = w1 - lr*w1b, w3 - lr*w3b
	w2, w4 = w2 - lr*w2b, w4 - lr*w4b
	b1, b2 = b1 - lr*b1b, b2 - lr*b2b
	w5, w7 = w5 - lr*w5b, w7 - lr*w7b
	w6, w8 = w6 - lr*w6b, w8 - lr*w8b
	b3, b4 = b3 - lr*b3b, b4 - lr*b4b

print('epoch = %d' %epoch)
print(' o1,  o2  = %6.3f, %6.3f' %(o1, o2))
```

## 2. 탄젠트(tanh) 함수

```python
from math import exp, tanh, log

i1, i2 = .05, .10
t1, t2 =   0,   1

w1, w3 = .15, .25
w2, w4 = .20, .30
b1, b2 = .35, .35

w5, w7 = .40, .50
w6, w8 = .45, .55
b3, b4 = .60, .60

for epoch in range(2000000):

	#print('epoch = %d' %epoch)

	h1 = i1*w1 + i2*w2 + 1*b1
	h2 = i1*w3 + i2*w4 + 1*b2
	h1 = tanh(h1) #h1노드에 순전파 탄젠트 활성화 함수 적용
	h2 = tanh(h2) #h2노드에 순전파 탄젠트 활성화 함수 적용

	o1 = h1*w5 + h2*w6 + 1*b3
	o2 = h1*w7 + h2*w8 + 1*b4
	o1m = o1 - max(o1, o2) #출력 층의 활성화 함수를 소프트맥스로 변경
	o2m = o2 - max(o1, o2) #출력 층의 활성화 함수를 소프트맥스로 변경
	o1 = exp(o1m)/(exp(o1m)+exp(o2m)) #출력 층의 활성화 함수를 소프트맥스로 변경
	o2 = exp(o2m)/(exp(o1m)+exp(o2m)) #출력 층의 활성화 함수를 소프트맥스로 변경

	#print(' o1,  o2  = %6.3f, %6.3f' %(o1, o2))

	E = -t1*log(o1) + -t2*log(o2)
	if E < 0.0001:              #for문을 빠져나가는 오차 값을 0.0001로 변경
		break

	o1b, o2b = o1 - t1, o2 - t2 #소프트맥스 함수의 역전파 오차 계산 부분
	# nothing for softmax + cross entropy error

	h1b, h2b = o1b*w5+o2b*w7, o1b*w6+o2b*w8
	h1b, h2b = h1b*(1+h1)*(1-h1), h2b*(1+h2)*(1-h2) #h1b, h2b노드에 역전파 탄젠트 활성화 함수 적용

	w1b, w3b = i1*h1b, i1*h2b
	w2b, w4b = i2*h1b, i2*h2b
	b1b, b2b = 1*h1b, 1*h2b
	w5b, w7b = h1*o1b, h1*o2b
	w6b, w8b = h2*o1b, h2*o2b
	b3b, b4b = 1*o1b, 1*o2b

	lr = 0.01
	w1, w3 = w1 - lr*w1b, w3 - lr*w3b
	w2, w4 = w2 - lr*w2b, w4 - lr*w4b
	b1, b2 = b1 - lr*b1b, b2 - lr*b2b
	w5, w7 = w5 - lr*w5b, w7 - lr*w7b
	w6, w8 = w6 - lr*w6b, w8 - lr*w8b
	b3, b4 = b3 - lr*b3b, b4 - lr*b4b

print('epoch = %d' %epoch)
print(' o1,  o2  = %6.3f, %6.3f' %(o1, o2))
```

## 3. 렐루(LeLU) 함수

```python
from math import exp, tanh, log

i1, i2 = .05, .10
t1, t2 =   0,   1

w1, w3 = .15, .25
w2, w4 = .20, .30
b1, b2 = .35, .35

w5, w7 = .40, .50
w6, w8 = .45, .55
b3, b4 = .60, .60

for epoch in range(2000000):

	#print('epoch = %d' %epoch)

	h1 = i1*w1 + i2*w2 + 1*b1
	h2 = i1*w3 + i2*w4 + 1*b2
	h1 = (h1>0)*h1 #h1노드에 순전파 렐루 활성화 함수 적용
	h2 = (h2>0)*h2 #h2노드에 순전파 렐루 활성화 함수 적용

	o1 = h1*w5 + h2*w6 + 1*b3
	o2 = h1*w7 + h2*w8 + 1*b4
	o1m = o1 - max(o1, o2) #출력 층의 활성화 함수를 소프트맥스로 변경
	o2m = o2 - max(o1, o2) #출력 층의 활성화 함수를 소프트맥스로 변경
	o1 = exp(o1m)/(exp(o1m)+exp(o2m)) #출력 층의 활성화 함수를 소프트맥스로 변경
	o2 = exp(o2m)/(exp(o1m)+exp(o2m))#출력 층의 활성화 함수를 소프트맥스로 변경

	#print(' o1,  o2  = %6.3f, %6.3f' %(o1, o2))

	E = -t1*log(o1) + -t2*log(o2)
	if E < 0.0001:            #for문을 빠져나가는 오차 값을 0.0001로 변경
		break

	o1b, o2b = o1 - t1, o2 - t2 #소프트맥스 함수의 역전파 오차 계산 부분
	# nothing for softmax + cross entropy error

	h1b, h2b = o1b*w5+o2b*w7, o1b*w6+o2b*w8
	h1b, h2b = h1b*(h1>0)*1, h2b*(h2>0)*1 #h1b, h2b노드에 역전파 렐루 활성화 함수 적용

	w1b, w3b = i1*h1b, i1*h2b
	w2b, w4b = i2*h1b, i2*h2b
	b1b, b2b = 1*h1b, 1*h2b
	w5b, w7b = h1*o1b, h1*o2b
	w6b, w8b = h2*o1b, h2*o2b
	b3b, b4b = 1*o1b, 1*o2b

	lr = 0.01
	w1, w3 = w1 - lr*w1b, w3 - lr*w3b
	w2, w4 = w2 - lr*w2b, w4 - lr*w4b
	b1, b2 = b1 - lr*b1b, b2 - lr*b2b
	w5, w7 = w5 - lr*w5b, w7 - lr*w7b
	w6, w8 = w6 - lr*w6b, w8 - lr*w8b
	b3, b4 = b3 - lr*b3b, b4 - lr*b4b

print('epoch = %d' %epoch)
print(' o1,  o2  = %6.3f, %6.3f' %(o1, o2))
```
