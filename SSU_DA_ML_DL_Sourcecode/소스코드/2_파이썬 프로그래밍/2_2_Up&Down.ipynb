##1. Up&Down

```python
import random  					   #랜덤 함수를 실행하기 위한 모듈

#찾을 숫자의 범위 내에서 임의의 자연수 정하기
final = int(input('찾을 자연수 범위 : 1 ~ ?'))		  #찾을 자연수의 범위
h = random.randint(1, final)			  #찾을 숫자를 h 변수에 저장
print('랜덤숫자 = ', h)				  #찾을 숫자를 출력

me = int(input('정답이라고 생각하는 자연수 : '))               #정답 자연수 입력


#정답 범위 출력
if(h > me) : print('UP!하세요. 정답보다 작습니다')            #찾을 숫자가 선택한 숫자보다 작으면 UP하라고 출력
elif(h < me) : print('DOWN!하세요. 정답보다 큽니다')     #찾을 숫자가 선택한 숫자보다 크면 DOWN하라고 출력
else : print('정답입니다')				  #찾을 숫자가 선택한 숫자와 같으면 정답이라고 출력
```

##2. Up&Down 재귀함수

```python
import random				 #랜덤 함수를 실행하기 위한 모듈

#찾을 숫자의 범위 내에서 임의의 자연수 정하기
final = int(input('찾을 자연수 범위 : 1 ~ ?'))		 #찾을 자연수의 범위
h = random.randint(1, final)			 #찾을 숫자를 h 변수에 저장
me = int(input('정답이라고 생각하는 자연수 : '))	 #정답 자연수 입력

def findNumber(num) : 			 #컴퓨터가 선택할 숫자를 정하는 함수
    global count				 #몇 번만에 찾았는지 저장하는 변수(전역변수)
    count+=1     				 #찾은 횟수 1증가

    if(h > num) :				 #찾을 숫자가 선택한 숫자보다 크면
        me = int(input('UP하세요. 자연수 : ')) 		 #UP하라고 출력하고 새로운 값 입력받기
        findNumber(me)				 #숫자를 선택하는 함수 호출
    elif(h < num) :				 #찾을 숫자가 선택한 숫자보다 작으면
        me = int(input('DOWN하세요. 자연수 : ')) 	 #DOWN하라고 출력하고 새로운 값 입력받기
        findNumber(me)				 #숫자를 선택하는 함수 호출
    else : 					 #찾을 숫자가 선택한 숫자와 같으면
        print('정답입니다')			 #정답이라고 출력
        print('시도 횟수 : ', count)			 #시도 횟수 출력
count = 0					 #초기 시도 횟수는 0으로 저장
findNumber(me)				 #숫자 찾는 함수 실행
```

##4. Up&Down 2씩 증가

```python
import random			 #랜덤 함수를 실행하기 위한 모듈

#찾을 숫자의 범위 내에서 임의의 자연수 정하기
final = int(input('찾을 자연수 범위 : 1 ~ ?'))	 #찾을 자연수의 범위
h = random.randint(1, final)		 #찾을 숫자를 h 변수에 저장
print('랜덤숫자 = ', h) 			 #찾을 숫자를 출력

#컴퓨터가 선택할 숫자 정하기
def findNumber(num) : 			 #컴퓨터가 선택할 숫자를 정하는 함수
    global count			 #몇 번만에 찾았는지 저장하는 변수(전역변수)
    count+=1			 #찾은 횟수 1증가
    print('#', count, '선택:', num)		 #선택한 숫자를 출력

    if(h > num) :			 #찾을 숫자가 선택한 숫자보다 크면
        print('UP!하세요. 정답보다 작습니다')	 #UP하라고 출력
        num+=2			 #선택한 숫자를 2증가
        findNumber(num)			 #숫자를 선택하는 함수 호출
    elif(h < num) : 			 #찾을 숫자가 선택한 숫자보다 작으면
        print('DOWN!하세요. 정답보다 큽니다')      #DOWN하라고 출력
        num-=1			 #선택한 숫자를 1감소
        findNumber(num) 		 #숫자를 선택하는 함수 호출
    else : 				 #찾을 숫자가 선택한 숫자와 같으면
        print('정답입니다')			 #정답이라고 출력
        print('시도 횟수 : ', count)		 #시도 횟수 출력
count = 0				 #초기 시도 횟수는 0으로 저장
findNumber(1)			 #숫자 찾는 함수 실행(시작은 1)

```

##5. Up&Down 평균

```python
import random #랜덤 함수를 실행하기 위한

#찾을 숫자의 범위 내에서 임의의 자연수 정하기
final = int(input('찾을 자연수 범위 : 1 ~ ?'))
h = random.randint(1, final)
print('랜덤숫자 = ', h)

#컴퓨터가 선택할 숫자 정하기
def findNumber(start, end) :
    global count
    count+=1
    m = start+round((end-start)/2)
    print('#', count, '선택:', m)

    if(h > m) :
        print('UP!하세요. 정답보다 작습니다')
        print('findNumber(', m,',', end,')')
        findNumber(m, end)
    elif(h < m) :
        print('DOWN!하세요. 정답보다 큽니다')
        print('findNumber(',start,',', m,')')
        findNumber(start, m)

    else :
        print('정답입니다')
        print('시도 횟수 : ', count)

#시도 횟수
count = 0
findNumber(1, final)
```
