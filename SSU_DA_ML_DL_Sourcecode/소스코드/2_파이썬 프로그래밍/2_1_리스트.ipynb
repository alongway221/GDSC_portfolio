```python
def josephus_permutation(n, k):
  people = list(range(1, n + 1))  		        #1~8미만
  result = []                                 #결과를 저장하는 리스트 초기화
  index = 0                                   #순서를 나타내는 변수

  while len(people) > 0:			               #리스트가 비어 있을 때까지 반복
    index = (index + k - 1) % len(people) 	 #리스트의 현재 인덱스를 계산
    result.append(people.pop(index))	       #해당 인덱스의 사람을 제거하여 결과 리스트에 추가

  return result

n = 7 #사람의 수
k = 3 #제거할 간격

print("요세푸스 순열:", josephus_permutation(n, k)) #요세푸스 순열 계산 함수 호출 및 출력
```

```python
#변수 선언
dic = {}        #영어 단어를 저장할 딕셔너리
answer = []     #입력한 값을 저장하는 리스트
correct = []    #정답을 저장하는 리스트
score = 0       #정답 개수를 저장하는 변수

#영어 단어 입력
for i in range(0,3):
    a = input('영어단어:')
    b = input('뜻')
    dic[a]=b

#영어 단어 입력
for i in dic.keys():
    c = input(i+'의 뜻:')
    answer.append(c)

#영어 단어 정답 저장
for i in dic.values():
    correct.append(i)

#영어 단어 정답 비교
for i in range(0,3):
    d = answer[i]
    e = correct[i]

    if d==e : score = score + 1

print(answer)
print(correct)
print('영어 시험 점수는',score,'점 입니다')
```

```python
def calculate_time(word):
    # 다이얼 알파벳에 해당하는 숫자 매핑
    dial = {
        'A': 2, 'B': 2, 'C': 2,
        'D': 3, 'E': 3, 'F': 3,
        'G': 4, 'H': 4, 'I': 4,
        'J': 5, 'K': 5, 'L': 5,
        'M': 6, 'N': 6, 'O': 6,
        'P': 7, 'Q': 7, 'R': 7, 'S': 7,
        'T': 8, 'U': 8, 'V': 8,
        'W': 9, 'X': 9, 'Y': 9, 'Z': 9
    }

    # 전화 거는 데 걸리는 시간 계산
    total_time = 0
    for char in word:
        total_time += dial[char] + 1  # 숫자 1을 거는 데 2초 걸리므로 +1 추가

    return total_time

# 예시 단어
word = "UNUCIC"
print("전화 거는 데 필요한 시간:", calculate_time(word))
```
