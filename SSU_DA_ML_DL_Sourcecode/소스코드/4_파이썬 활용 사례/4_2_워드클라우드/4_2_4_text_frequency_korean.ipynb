```python
from konlpy.tag import Twitter
nlpy = Twitter()
nouns = nlpy.nouns(lines)
print(nouns)
```

```python
from collections import Counter
count = Counter(nouns)

tag_count = []
tags = []

for n, c in count.most_common(100):
 dics = {'tag': n, 'count': c}

 if len(dics['tag']) >= 2 and len(tags) <= 49:
  tag_count.append(dics)
  tags.append(dics['tag'])
```

```python
for tag in tag_count:
 print(" {:<14}".format(tag['tag']), end='\t')
 print("{}".format(tag['count']))
```

```python
word_list = []

for tag in tag_count:
 b = tag['tag'] + ' '
 word_list.append(b*tag['count'])

#print(word_list)
a = ' '.join(s for s in word_list)
print(a)
```

```python
img = Image.open('heart.png') #추가
img_array = np.array(img)     #추가

wordcloud = WordCloud(font_path="NanumGothicExtraBold.ttf", width=400, height=400, background_color="white", mask=img_array)
wordcloud = wordcloud.generate(a)

plt.figure() #추가
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
