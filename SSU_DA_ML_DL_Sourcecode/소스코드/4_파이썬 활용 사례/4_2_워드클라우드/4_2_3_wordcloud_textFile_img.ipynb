```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
#a = '저는 수학을 참 좋아합니다. 수학 문제를 풀 때'
with open('bts.txt', 'r', encoding='utf-8') as f:
    a = f.read()

img = Image.open('heart.png') #추가
img_array = np.array(img)     #추가

wordcloud = WordCloud(font_path="NanumGothicExtraBold.ttf", width=400, height=400, background_color="white", mask=img_array)
wordcloud = wordcloud.generate(a)

plt.figure() #추가
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
