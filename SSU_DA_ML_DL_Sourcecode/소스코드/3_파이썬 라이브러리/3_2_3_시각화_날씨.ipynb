```python
#Pandas 라이브러리를 pd 라는 이름으로 가져오기
import pandas as pd		

#pd에 있는 read_csv()함수로 .csv 파일 읽기
df = pd.read_csv ('./weather.csv', encoding='utf-8') 
```

```python
df
```

```python
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize=(10,5)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(df['high']) ## 선그래프 생성
ax.plot(df['low']) ## 선그래프 생성
ax.plot(df['average'], color='#ff0000') ## 선그래프 생성
 
plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 40도 회전 
plt.title('weather graph',fontsize=20) ## 타이틀 설정
plt.show()
```

```python
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize=(10,5)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성

#그래프 꾸미기 
ax.plot(df['high'], color='#ff0000', linestyle='--',linewidth=2, marker='o',markersize=10, markeredgecolor='red', markeredgewidth=3)
ax.plot(df['low'], color='#00ff00', linestyle='dashdot',linewidth=2, marker='o',markersize=10, markeredgecolor='green', markeredgewidth=3)
ax.plot(df['average'], color='#0000ff', linestyle='solid', linewidth=2, marker='o',markersize=10, markeredgecolor='blue', markeredgewidth=3) 
 
plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 40도 회전 
plt.title('weather graph',fontsize=20) ## 타이틀 설정
plt.show()
```
