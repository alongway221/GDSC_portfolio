```python
import pandas as pd
 
data = {
    'year': [2016, 2017, 2018],
    'GDP rate': [2.8, 3.1, 3.0],
    'GDP': ['1.637M', '1.73M', '1.83M']
}
 
df = pd.DataFrame(data)
print(df)
```

```python
print(df['year']) #년도만 출력
```

```python
print(df.year)   #이렇게 사용도 가능
```

```python
df[df.year>2016] #년도가 2016보다 큰 것만 출력
```

```python
sum = df['GDP rate'].sum()   #GDP rate의 합
avg = df['GDP rate'].mean()  #GDP rate의 평균
print(sum, avg)
```
